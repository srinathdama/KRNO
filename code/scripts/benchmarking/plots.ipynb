```python
import numpy as np
import matplotlib.pyplot as plt


```

```python


# Data from the tables
quadrature_nodes = np.array([1, 4, 9, 16, 25]) * (32 * 32 * 5)  # Convert N to actual number of nodes   
krno_train_time = np.array([0.0279, 0.0394, 0.0626, 0.0999, 0.1584])
fno3d_train_time = np.array([0.0216, 0.0252, 0.0405, 0.0704, 0.1114])
krno_test_time = np.array([0.0107, 0.0134, 0.0185, 0.0312, 0.0502])
fno3d_test_time = np.array([0.0065, 0.0069, 0.0108, 0.0210, 0.0315])
```

```python
import pathlib
import sys, os

CURR_DIR = pathlib.Path(' ').parent.absolute()
FIG_DIR = CURR_DIR / "figs"
os.makedirs(FIG_DIR, exist_ok=True)

sys.path.append(str(CURR_DIR / ".."))
CURR_DIR
```

```python
from plotting_config import PlotConfig
PlotConfig.setup()

figsize = PlotConfig.convert_width((10, 6), page_scale=1.0)
fig, ax = plt.subplots(figsize=figsize)

# Plot training and testing times in a single plot
# plt.figure(figsize=(10, 6))

# Training times
plt.plot(quadrature_nodes, krno_train_time, 'ko-', label="KRNO (train)")
plt.plot(quadrature_nodes, fno3d_train_time, 'bo-', label="FNO-3D (train)")

# Testing times
plt.plot(quadrature_nodes, krno_test_time, 'k*--', label="KRNO (test)")
plt.plot(quadrature_nodes, fno3d_test_time, 'b*--', label="FNO-3D (test)")

plt.xlabel("No of points in spatio-temporal grid", fontsize=12)
plt.ylabel("Time (seconds)", fontsize=12)
# plt.title("Training and Testing Time vs Quadrature Nodes", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
PlotConfig.save_fig(fig, str(FIG_DIR / 'timing'))
```
